```python
#Voting

ridge = pyc.create_model('ridge')
logistic = pyc.create_model('lr')
lightgbm = pyc.create_model('lightgbm')
```

```python
voting_model = pyc.blend_models([ridge,logistic,lightgbm])
```

```python
voting_model
```

```python
knn_model = pyc.create_model('knn')
```

```python
stacking_model = pyc.stack_models([ridge,logistic,lightgbm], meta_model = ridge)
```

```python
stacking_model
```

```python
pyc.evaluate_model(stacking_model)
```

```python
test = pd.read_csv(test_path)
```

```python
test
```

```python
test_prediction = pyc.predict_model(stacking_model,data = test,raw_score = True)[['Loan_ID','prediction_label']]
```

```python
test_prediction.columns = ['Loan_ID','Loan_Status']
```

```python
test_prediction.to_csv('loan_result_stacking.csv', index = False)
```

```python
stacking_final_model_pipeline = pyc.finalize_model(stacking_model)
```

```python
stacking_final_model_pipeline
```

```python
pyc.save_model(stacking_final_model_pipeline, 'stacking_final_model_pipeline')
```

```python
loaded_model = pyc.load_model('stacking_final_model_pipeline')
```

```python
pyc.predict_model(loaded_model,data = test,raw_score = True)[['Loan_ID','prediction_label']]
```
